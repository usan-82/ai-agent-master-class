```python
####################### Memorys & Tools  ##################
## <-- Memory만 적용 시 챗봇수준
# 이 챗봇은 우리 시스템 안에 실행할 수 있는 함수들이 있다는 개념 자체를 모르기 때문
# AI에게 API로 요청 시 우리 시스템엔 이런 함수/기능들이 있다 라는 지침을 미리 내려두면 
# 서버에서 알아서 처리해줄것임 
# 이를 위해 Tools 리스트를 생성함
```

```python
import openai, json

client = openai.OpenAI()
messages = []
```

```python
def get_weather(city):
    return "33 degrees celcius."

FUNCTION_MAP = {'get_weather': get_weather}
```

```python
TOOLS = [
    {
        "type":"function",
        "function": {
            "name": "get_weather",
            "description": "A function to get the weather of a city.",
            "parameters":{
                "type":"object",
                "properties":{
                    "city":{
                        "type:":"string",
                        "description":"The name of the city to get the weather of."
                    }
                },
                "required":["city"]
            }
        }
    }
]
```

```python
from openai.types.chat import ChatCompletionMessage

def process_ai_response(message: ChatCompletionMessage):
    #ai가 대화의 전체 히스토리를 볼 수 있게 하고자 함
    if message.tool_calls:
        messages.append(
            {
                "role": "assistant",
                "content": message.content or "",
                #tool_call리스트에 있는 tool_call 개수만큼 동적으로 object를 만들어냄 
                "tool_calls": [{
                    "id": tool_call.id,
                    "type": "function",
                    "function": {
                        "name": tool_call.function.name,
                        "arguments": tool_call.function.arguments,
                    },
                } for tool_call in message.tool_calls]
            } # 헤이, 나는 기록을 저장하기를 원해. ai 네가 나한테 이 함수를 arguments와 함께 실행해달라고 한 것을 말야
        )

        #함수 실행
        for tool_call in message.tool_calls: 
            function_name = tool_call.function.name
            arguments = tool_call.function.arguments

            print(f"Calling function: {function_name} with {arguments}") #ai모델이 주는 arguments는 '{"key":"value"}' 로, json형태의 "문자열"

            #arguments로 받은 json형태의 문자열을 Python 객체(dictionary)로 변경
            try:
                arguments = json.loads(arguments)
            except json.JSONDecodeError:
                arguments = {}

            function_to_run = FUNCTION_MAP.get(function_name)

            # **의 역할: {"key":"value"} 형태의 딕셔너리를 key="value" 형태로 변경해줌
            result = function_to_run(**arguments)
            print(f"Ran {function_name} with args {arguments} for a result of {result}")

            messages.append(
                {
                    "role": "tool",
                    "tool_call_id": tool_call.id,
                    "name": function_name,
                    "content": result, 
                }
            )

        call_ai()
    else: # ai가 일반 텍스트로 대화
        messages.append({"role":"assistant","content":message.content})
        print(f"AI: {message.content}")


#############################################################

def call_ai():
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        tools=TOOLS
    )
    process_ai_response(response.choices[0].message)
    print(response) #모델이 함수를 호출하고싶을 때 작동방식 확인가능
    message = response.choices[0].message.content
    messages.append({"role":"assistant","content":message})
    print(f"AI:{message}")
```

```python
while True:
    message = input("LLM에게 메세지를 보내주세요 : ")
    if message == "대화종료" or message == "q":
        break
    else:
        messages.append({"role":"user","content":message})
        print(f"User:{message}")
        call_ai()
```
